This notebook creates a group bar chart with labels comparison of 3 datasets

```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
isic_df = pd.read_excel('ISIC_2019_ita_tone_stats.xlsx',index_col=0)  
print(isic_df.Counts.sum())
isic_df = isic_df.reset_index()
custom_dict = {'Very light':0, 'Light':1, 'Intermediate':2, 'Tan':3, 'Brown':4, 'Dark':5, 'Undetermined': 6}
isic_df['index'] = isic_df['Skin tone'].map(custom_dict)
isic_df = isic_df.set_index('index')
isic_df = isic_df.sort_index()
isic_df = isic_df.set_index('Skin tone')
isic_df
```

```python
sd_df = pd.read_excel('SD-260_ita_tone_stats.xlsx',index_col=0)  
print(sd_df.Counts.sum())
sd_df = sd_df.reset_index()
sd_df['index'] = sd_df['Skin tone'].map(custom_dict)
sd_df = sd_df.set_index('index')
sd_df = sd_df.sort_index()
sd_df = sd_df.set_index('Skin tone')
sd_df
```

```python
ham_df = pd.read_excel('HAM10000_ita_tone_stats.xlsx',index_col=0)  
print(ham_df.Counts.sum())
ham_df = ham_df.reset_index()
ham_df['index'] = ham_df['Skin tone'].map(custom_dict)
ham_df = ham_df.set_index('index')
ham_df = ham_df.sort_index()
ham_df = ham_df.set_index('Skin tone')
ham_df
```

```python
#isic_df.Counts.sum()/ sd_df.Counts.sum()
```

```python
isic_df['Proportion'] = isic_df.Counts.apply(lambda x: round(100*x/sum(isic_df.Counts),2))
isic_df
```

```python
sd_df['Proportion'] = sd_df.Counts.apply(lambda x: round(100*x/sum(sd_df.Counts),2))
sd_df
```

```python
ham_df['Proportion'] = ham_df.Counts.apply(lambda x: round(100*x/sum(ham_df.Counts),2))
ham_df
```

```python
x1 = sd_df.index
y1 = sd_df.Proportion

x2 = isic_df.index
y2 = isic_df.Proportion

x3 = ham_df.index
y3 = ham_df.Proportion
```

```python
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
labels = x1
sd_means = y1
isic_means = y2
ham_means = y3

x = np.arange(len(labels))  # the label locations
width = 0.3  # the width of the bars

# Grab only the colors that appear
colors = []
mycolors = {'Very light':'antiquewhite', 'Light':'bisque', 'Intermediate':'burlywood', 'Tan':'xkcd:puce', 'Brown':'xkcd:brown', 'Dark':'xkcd:dark brown', 'Undetermined': 'white'}
for key in mycolors.keys():
    if key in labels:
        colors.append(mycolors[key])

fig, ax = plt.subplots(figsize=(12,5))
rects1 = ax.bar(x - width, sd_means, width, label='SD-260', color=colors, edgecolor='blue')
rects2 = ax.bar(x, isic_means, width, label='ISIC 2019',color=colors, edgecolor='red')
rects3 = ax.bar(x + width, ham_means, width, label='HAM10000',color=colors, edgecolor='purple')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Images percentages')
ax.set_xlabel('Skin tones')
#ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()
fig.savefig(f'congress_all_skin_tones_bars.png', dpi=300,bbox_inches="tight")
```
